```python
def get_list_of_major_cities():  
    website = 'https://www.infoplease.com/us/us-cities/top-50-cities-us-population-and-rank'
    data = pd.read_html(website, index_col=0)
    city = pd.DataFrame()
    city = city.append(data[0], ignore_index=True)
    city.columns = ['Cities','pop1','pop2','pop3','pop4','pop5','pop6']
    city = city.drop([0], axis=0)
    city = city.drop(['pop1','pop2','pop3','pop4','pop5','pop6'], axis=1)
    return city
```

```python
def cityweather(cityname):
    key='89534505a9b84568b10223850183011'
    url='https://api.worldweatheronline.com/premium/v1/weather.ashx'
    today = datetime.datetime.now().strftime("%Y-%m-%d")
    p = {'q': cityname, 'date' : 2017-3-1,'enddate':2017-3-31, 'format' : 'json', 'key' : key}
    response = requests.get(url, params=p).json()
    return response
```

```python
cities = get_list_of_major_cities()

max_temps = []
min_temps = []
rainfall = []
for city in cities['Cities']:
    print(city)
    response = cityweather(city)
    total_yearlyminF = 0
    total_yearlymaxF = 0
    total_yearlyrainfall = 0
    for month in response['data']['ClimateAverages'][0]['month']:
        total_yearlyminF += float(month['avgMinTemp_F'])
        total_yearlymaxF += float(month['absMaxTemp_F'])
        total_yearlyrainfall += float(month['avgDailyRainfall'])
    total_yearlyminF = total_yearlyminF/12
    total_yearlymaxF = total_yearlymaxF/12
    total_yearlyrainfall = total_yearlyrainfall/12
    max_temps.append(total_yearlymaxF)
    min_temps.append(total_yearlyminF)
    rainfall.append(total_yearlyrainfall)

cities['AvgYear_Max_Temp'] = pd.Series(data = max_temps, index = cities.index)
cities['AvgYear_Min_Temp'] = pd.Series(data = min_temps, index = cities.index)
cities['AvgYear_Rainfall'] = pd.Series(data = rainfall, index = cities.index)

biomes = []
for index, row in cities.iterrows():
    avg_temp = (row['AvgYear_Max_Temp'] + row['AvgYear_Min_Temp'])/2
    avg_rainfall = row['AvgYear_Rainfall']
    if (avg_temp >= 68 and avg_temp <= 85 and avg_rainfall >= 0.456712328767123 and avg_rainfall <= 1.295890410958904):
        biomes.append("Desert")
    elif (avg_temp >= 42.5 and avg_temp <= 75 and avg_rainfall >= 1.391780821917808 and avg_rainfall <= 3.435616438356164):
        biomes.append("Grasslands")
    elif (avg_temp >= 50 and avg_temp <= 82.5 and avg_rainfall >= 2.087671232876712 and avg_rainfall <= 8.175342465753425):
        biomes.append("Deciduous/Temperate Forest")
    elif (avg_temp >= 37.3 and avg_temp <= 60 and avg_rainfall >= 1.391780821917808 and avg_rainfall <= 3.479452054794521):
        biomes.append("Savanna")
    elif (avg_temp >= 82.6 and avg_temp <= 95 and avg_rainfall >= 9.479452054794521 and avg_rainfall <= 18.093150684931507):
        biomes.append("Rainforest")
    elif (avg_temp >= 10 and avg_temp <= 55 and avg_rainfall >= 0.69041095890411 and avg_rainfall <= 1.404109589041096):
        biomes.append("Alpine Tundra")
    else:
        biomes.append("Other")
cities['Biome'] = pd.Series(data = biomes, index = cities.index)
print(cities)
```
